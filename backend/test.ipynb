```python
# Retrival Of data from MongoDB 
from pymongo import MongoClient
import pandas as pd



client = MongoClient("DB URL With Password", tlsCAFile=certifi.where())

# Database and Collection name
db = client["MajorProject_Test"]
collection = db["sales_data"]


cursor = collection.find()

data = list(cursor)



df = pd.json_normalize(data)

# Remove MongoDB ObjectId column if not needed
if "_id" in df.columns:
    df.drop("_id", axis=1, inplace=True)

print("Data loaded into DataFrame")
print(df.head())

print("END")

```

```python
# To Insert Any CSV File to Supabase For testing purposes ONLY

import pandas as pd
from sqlalchemy import create_engine

# -----------------------------
# 1. Supabase Database Config
# -----------------------------

DB_USER = "postgres"
DB_PASSWORD = "YOUR_PASSWORD"
DB_HOST = "db.xxxxxxxxx.supabase.co"
DB_PORT = "5432"
DB_NAME = "postgres"

DATABASE_URL="DB URL With Password"
# Create SQLAlchemy engine
engine = create_engine(DATABASE_URL)

# -----------------------------
# 2. Load CSV File
# -----------------------------

csv_file = "test2.csv"

df = pd.read_csv(csv_file)

print("CSV Loaded Successfully")
print(df.head())

# -----------------------------
# 3. Upload Data to Supabase
# -----------------------------

table_name = "sales_data"

df.to_sql(
    table_name,
    engine,
    if_exists="replace",  # options: replace / append
    index=False
)

print(f"CSV uploaded successfully to Supabase table: {table_name}")
```
